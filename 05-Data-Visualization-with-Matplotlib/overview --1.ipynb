```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
```

```python
x = np.linspace(0,5,11)
y = x**2
```

```python
x
```

```python
y
```

```python
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Title')
```

```python
# Creates blank canvas
fig = plt.figure()

axes1 = fig.add_axes([.1,.1,.8,1]) # main axes
axes2 = fig.add_axes([0.2,0.5,0.2,0.5])


# Larger Figure Axes 1
axes1.plot(x, y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 2 Title')

axes2.plot(x,y,'y-')

```

```python
fig,axes = plt.subplots(nrows=1, ncols=2)
```

```python
axes
```

```python
axes[0].plot(x,y)
axes[1].plot(y,x)

fig
```

```python
fig = plt.figure(figsize=(8,2))
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,label= 'First',linewidth=20)
ax.plot(y,x,label = 'Second')

ax.legend()
ax
```
